```python
#Vamos importar uma por uma e começar a ajustar elas
dim_cliente = pd.read_excel(arquivo_dimensoes, sheet_name = 'Clientes')
dim_funcionario= pd.read_excel(arquivo_dimensoes, sheet_name = 'Funcionarios')
dim_escritorio = pd.read_excel(arquivo_dimensoes, sheet_name = 'Escritorios')
dim_categoria = pd.read_excel(arquivo_dimensoes, sheet_name = 'Categoria')
dim_produto = pd.read_excel(arquivo_dimensoes, sheet_name = 'Produtos')
dim_fornecedor = pd.read_excel(arquivo_dimensoes, sheet_name = 'Fornecedores')
```

```python
#trabalhando a tabela dim_cliente
dim_cliente
```

Precisamos limpar o Fax e o Telefone para podermos trabalhar melhor 

```python
#Trocando os caracteres especiais das colunas de telefone e Fax 
dim_cliente['Fax'].replace("[^a-zA-Z0-9]", '', regex=True, inplace =True)
dim_cliente['Telefone'].replace("[^a-zA-Z0-9]", '', regex=True, inplace =True)

dim_cliente.head()
```

```python
#Vamos ver a tipagem das colunas para ve se estão de acordo com o que deveriam ser. E já verificar os nulos 
dim_cliente.info()
```

```python
# Como teremos valores nulos dentro do dataset em fax, telefone e CEP que são variaveis de tipagem inteira, que não possuem casa decimal, irei preenche-los com 0 
# Vamos mudar a tipagem para numeric das colunas de Fax, telefone e Cep para int, pois CEP não possui casa decimal
for atributo in ['Fax', 'Telefone','CEP']:
    #preenchendo com 0 os nulos das colunas
    dim_cliente[atributo] = dim_cliente[atributo].fillna(0)
    dim_cliente[atributo] = pd.to_numeric(dim_cliente[atributo], errors= 'ignore', downcast='integer')
```

```python
#Checando os valore alterados.
dim_cliente.info()
```

Vamos começar inserir elas no banco de dados SQLite3


```python
#Vamos criar um banco de dados com o nome de Refera
#Setando o caminho fora da pasta de ETL onde estará o arquivo do SQLite
database_name = 'refera'
conn = sql.connect('..\\banco_de_dados\\refera.db')

#Com o banco de dados criado vamos inserir suas dimensões 
```

Irei fazer Separado a insersão das tabelas para melhor manutenção caso queira apenas refazer uma das tabelas 

```python
#INSERINDO NO BANCO DE DADOS A TABELA DIM_CLIENTE
dim_cliente.to_sql('DIM_CLIENTE', conn, if_exists='replace', index= False)
```

```python
#INSERINDO NO BANCO DE DADOS A TABELA DIM_FUNCIONARIO
dim_funcionario.to_sql('DIM_FUNCIONARIO', conn, if_exists='replace', index= False)
```
